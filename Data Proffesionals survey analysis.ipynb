### Duplicating the dataset before cleaning

```python
#making copy of dataset

df = data_pro_df.copy()
```

#### Some columns are not neccessary for my analyst, so I will be deleting these columns:
    
   * Email
   * Date Taken (America/New_York)
   * Time Taken (America/New_York)
   * Browser
   * OS
   * City
   * Country
   * Referrer
   * Time Spent
   * Q13 - Ethnicity
   * Q6 - How Happy are you in your Current Position with the following? (Coworkers)
   * Q6 - How Happy are you in your Current Position with the following? (Upward Mobility)	
   * Q6 - How Happy are you in your Current Position with the following? (Learning New Things)

```python
#Dropping columns I wont be needing for my analysis

#list of columns to be dropped
cols = ['Email', 'Date Taken (America/New_York)', 'Time Taken (America/New_York)',
        'Browser', 'OS', 'City', 'Country', 'Referrer', 'Time Spent', 'Q13 - Ethnicity',
        'Q6 - How Happy are you in your Current Position with the following? (Coworkers)',
        'Q6 - How Happy are you in your Current Position with the following? (Upward Mobility)',
        'Q6 - How Happy are you in your Current Position with the following? (Learning New Things)']

for col in cols:
    df = df.drop(col, axis = 1)
```

### For the 'Current role', I want to standardize their answers and change 'Other Please specify' to just 'Others'

```python
#spliting the Current Role column suing '('

other_role = df['Q1 - Which Title Best Fits your Current Role?'].str.split('(', expand = True)
```

```python
#replacing the answers with the first part of the split

df['Q1 - Which Title Best Fits your Current Role?'] = other_role[0].str.strip()
```

```python
#confirm change

df.head()
```

```python
#changing 'Other' to 'Others'

df['Q1 - Which Title Best Fits your Current Role?'] = df['Q1 - Which Title Best Fits your Current Role?'].str.replace('Other', 'Others')

#confirm change

df.head()
```

### For the Salary I want to split the column, convert it to integer and get average salary from that

```python
#splitting the salary column by '-' and remove the 'k' & '+'

#split salary column on '-' and remove the 'k'
salary = df['Q3 - Current Yearly Salary (in USD)'].str.replace('k', '').str.split('-', expand = True).fillna(np.NaN)

#removing the '+'
salary[0] = salary[0].str.replace('+', '')

#fill the null rows with '225' for rows with '225+' initially so I can get the average
salary[1] = salary[1].fillna(salary[0])
```

```python
#convert the salaries from object to int so I can erform mathematical operations

cols = [0, 1]

for col in cols:
    salary[col] = salary[col].astype(int)
    

#get the average of the salaries
salary['avg'] = (salary[0] + salary[1]) / 2

#view the new salary subset
salary.head()
```

```python
#assign the avg salary to its own column in the main dataset

df['Average Salary'] = salary['avg']

#view dataset
df.head()
```

### I will be extracting all corrected spelled countries from the country column

```python
#standardize all answers as proper case for consistency

df['Q11 - Which Country do you live in?'] = df['Q11 - Which Country do you live in?'].str.title()

#removing the 'Other (Please Specify):'
country = df['Q11 - Which Country do you live in?'].str.split(':', expand = True).fillna(np.NaN)

#view data
country.head()
```

```python
#Filling the null values with the country names from the left side
country[1] = country[1].fillna(country[0])

#confirm change
country.head()
```

```python
#I used the new country column to fill/replace the countries in the main data
df['Q11 - Which Country do you live in?'] = country[1]

#view dataset
df.head()
```

### I will be changing the 'Others (Please Specify)' to just 'Others' for the Industry the participants worked in since its not relevant to me for this analysis

```python
#splitting column on '('
industry = df['Q4 - What Industry do you work in?'].str.split('(', expand = True)

#view subset data
industry
```

```python
#replace the new answers in the main dataset and change 'Other' to 'Others'
df['Q4 - What Industry do you work in?'] = industry[0].str.strip().replace('Other', 'Others')

df.head()
```

```python
df['Q4 - What Industry do you work in?'].value_counts()
```

### I will changing the open ended answers for Q8 to just 'Others'

```python
#splitting the column on '('
q8 = df['Q8 - If you were to look for a new job today, what would be the most important thing to you?'].str.split('(', expand = True)

#view subset
q8
```

```python
#replacing the new answers in the main data and changing 'Other' to 'Others'

df['Q8 - If you were to look for a new job today, what would be the most important thing to you?'] = q8[0].str.strip().replace('Other', 'Others')

#confirm change
df.head()
```

### I will be filtering just the countries with the correct spelling, I will do this by importing a dataset with the list of countries worldwide and the I will put the names of the countries in a list, its the list I will use as the criteria for filtering.

```python
#importing the new dataset with the list of countries
countries = pd.read_csv('countries.csv')

#view dataset
countries
```

```python
#putting the names of the countries in a list
country = []
num = countries.shape[0]
for i in range(0, num):
    country.append(countries['Countries'][i])

#view list
country
```

```python
#filtering the dataset using the names of countries
df = df[df['Q11 - Which Country do you live in?'].isin(country)]
```
